## Extracting/Formatting the datas

```python

dataset = pd.read_csv('/Users/louise_monciero/Desktop/APU/Machine_Learning/Assignement/out copie.csv')

# Balancing the entries based on TV (=['label'])
phishing_sample = dataset[dataset['label'] == 'phishing'].sample(n=5000, random_state=42)
legitimate_sample = dataset[dataset['label'] == 'legitimate'].sample(n=5000, random_state=42)
dataset_sample = pd.concat([phishing_sample, legitimate_sample])

dataset_sample.drop(columns=['source'])
```
